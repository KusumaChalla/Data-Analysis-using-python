# Seaborn

### Main components of Seaborn

<h3 style= 'color:red'>Figure-level interface<h3> 
    
The relplot() function is an example of the figure-level interface. It can be used to create a figure consisting of multiple subplots, each showing the relationship between two variables.
    

```python
import seaborn as sns
tips = sns.load_dataset("tips")
sns.relplot(x="total_bill", y="tip", hue="smoker", col="time", data=tips)
```

This creates a figure with two subplots, one for dinner and one for lunch, each showing the relationship between total_bill and tip, colored by smoker.

<h3 style= 'color:red'>Axes-level interface<h3> 

The scatterplot() function is an example of the axes-level interface. It can be used to create a scatter plot of two variables. 

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
ax = sns.scatterplot(x="total_bill", y="tip", hue="smoker", data=tips)
ax.set_title("Total Bill vs. Tip")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Tip")
sns.despine()
plt.show()
```

This creates a scatter plot of total_bill versus tip, colored by smoker

<h3 style= 'color:red'>Themes and color palettes<h3>

Seaborn includes several built-in themes and color palettes that can be used to customize the look and feel of your visualizations. 

```python
import seaborn as sns
sns.set_style("darkgrid")
tips = sns.load_dataset("tips")
sns.scatterplot(x="total_bill", y="tip", hue="smoker", data=tips)
plt.show()
```

- This sets the theme to "darkgrid", which gives a dark gray background with gridlines. 
- You can also use the set_palette() function to change the color palette used in your plots.

<h3 style= 'color:red'>Statistical estimation:<h3>
Seaborn provides several functions for estimating statistical relationships between variables.

```python
import seaborn as sns
tips = sns.load_dataset("tips")
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips)
plt.show();
```

This creates a scatter plot of total_bill versus tip, colored by smoker, and fits a linear regression model to the data.

<h3 style= 'color:red'>Plotting functions:<h3>
Seaborn provides a wide range of plotting functions for creating different types of plots

```python
import seaborn as sns
flights = sns.load_dataset("flights")
flights = flights.pivot("month", "year", "passengers")
sns.heatmap(flights, annot=True, fmt="d", cmap="YlGnBu")
plt.show()
```

<h3 style= 'color:red'>Matrix plots<h3>
Matrix plots are used to visualize the relationships between multiple variables, usually by displaying a matrix of values or a heatmap

#### some examples of how to use these plot matrix plots in Seaborn

__Heatmap__

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

corr = tips.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

__Clustermap__:

```python
import seaborn as sns

tips = sns.load_dataset("tips")

plt.figure(figsize=(2,2), dpi=40)
g = sns.clustermap(tips.corr(), annot=True, cmap="coolwarm")
g.ax_row_dendrogram.remove()
g.ax_col_dendrogram.remove()
plt.show()

```

__Pairplot__

```python
import seaborn as sns

tips = sns.load_dataset("tips")

sns.pairplot(tips, hue="sex", diag_kind="kde")
plt.show()
```

__Jointplot__

```python
import seaborn as sns

tips = sns.load_dataset("tips")

sns.jointplot(x="total_bill", y="tip", data=tips, kind="kde")
plt.show()
```

__Scatterplot__

```python
import seaborn as sns

tips = sns.load_dataset("tips")

sns.scatterplot(x="total_bill", y="tip", hue="sex", data=tips)
plt.show()
```

### To save the plot in Seaborn, we can use the savefig() function from the matplotlib library.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

g = sns.clustermap(tips.corr(), annot=True, cmap="coolwarm")
g.ax_row_dendrogram.remove()
g.ax_col_dendrogram.remove()

# save the plot as a PNG file
plt.savefig("my_plot.png", dpi=300)

# plt.show()
```
