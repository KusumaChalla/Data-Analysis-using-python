# Matplotlib

## Two main ways to create plots using Matplotlib: the functional method and the object-oriented method.

- ### Functional Method: 
    In the functional method, you use the pyplot module to create and customize plots. This approach is quick and easy for creating simple plots, but can become messy and harder to maintain when creating complex plots.


### Example:

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Create a line plot
plt.plot(x, y)

# Add labels and title
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Title of the plot')

# Show the plot
plt.show()
```

- ### Object-Oriented Method: 
    In the object-oriented method, you create a figure object and one or more axes objects, and then use methods of these objects to create and customize the plot. This approach is more flexible and scalable for creating complex and multi-axis plots.

### Example:

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Create a figure and axis object
fig, ax = plt.subplots()

# Create a line plot
ax.plot(x, y)

# Add labels and title
ax.set_xlabel('X-axis label')
ax.set_ylabel('Y-axis label')
ax.set_title('Title of the plot')

# Show the plot
plt.show()
```

### COMMON TYPES OF PLOTS THAT CAN BE CREATED USING PYPLOT:

### _Line Plot_:

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Temperature over Time')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sine Wave')
plt.show()
```

### _Scatter Plot_:

```python
import matplotlib.pyplot as plt

height = [62, 65, 73, 60, 70, 65, 75, 67, 72]
weight = [120, 150, 170, 110, 160, 130, 180, 140, 170]

plt.scatter(height, weight)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height vs. Weight')
plt.show()
```

### _Bar Plot_:

```python
import matplotlib.pyplot as plt

products = ['Product A', 'Product B', 'Product C']
sales = [100, 200, 150]

plt.bar(products, sales)
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Sales of Different Products')
plt.show()
```

### Histogram:

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(123)
ages = np.random.normal(40, 10, 100)

plt.hist(ages, bins=10)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Ages')
plt.show()
```

### _Pie Chart_:

```python
import matplotlib.pyplot as plt

sizes = [40, 20, 30, 10]
labels = ['Group A', 'Group B', 'Group C', 'Group D']

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Proportion of Each Group')
plt.show()
```

### _Box Plot_:

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(123)
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(70, 20, 100)
data3 = np.random.normal(90, 30, 100)
data = [data1, data2, data3]

plt.boxplot(data)
plt.xticks([1, 2, 3], ['Group 1', 'Group 2', 'Group 3'])
plt.xlabel('Group')
plt.ylabel('Value')
plt.title('Distribution of Values in Different Groups')
plt.show()
```

### _Area Plot_:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 100)
y1 = np.exp(-x**2)
y2 = np.exp(-(x-1)**2)
y = np.cumsum(y1 - y2)

plt.fill_between(x, y1, y2, color='gray', alpha=0.5)
plt.plot(x, y, color='black')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.title('Area Between Two Probability Distributions')
plt.show()
```

### _Heatmap_:

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.rand(10, 10)
plt.imshow(data, cmap='coolwarm')
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Heatmap')
plt.show()
```

### Contour Plot:

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
plt.contour(X, Y, Z)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Contour Plot')
plt.show()
```

### _3D Plot_:

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.random.randn(100)
y = np.random.randn(100)
z = np.random.randn(100)
ax.scatter(x, y, z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.title('3D Scatter Plot')
plt.show()
```

### violin plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some random data
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Create a violin plot
plt.violinplot(data)

# Add x label, y label, and title
plt.xlabel('Group')
plt.ylabel('Value')
plt.title('Violin Plot')

# Show the plot
plt.show()
```

### _Density Plot_:

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(1000)
plt.hist(data, density=True, alpha=0.5, bins=30)
plt.plot(np.linspace(-4, 4, 100), 1/np.sqrt(2*np.pi) * np.exp(-0.5*np.linspace(-4, 4, 100)**2), 'r')
plt.show()
```

### _Hexbin Plot_:

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.randn(1000)
y = np.random.randn(1000)
plt.hexbin(x, y, gridsize=20, cmap='YlOrBr')
plt.colorbar()
plt.show()
```
